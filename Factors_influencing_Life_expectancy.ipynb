# Bivarite Analysis

  This annalysis study the relationship between two columns.the purpose is to check if one variable affects or relates to another.
  particulary how different socio economic and health indicators affect life expectancy.


**Scatter Plot**

Next, let's explore the relationship between 'Life expectancy ' and 'GDP'.

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='GDP', y='Life expectancy ', data=df)
plt.title('Life Expectancy vs. GDP')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

The scatter plot indicates a positive correlation between GDP and Life expectancy, where higher GDP is generally associated with higher life expectancy.

this shows the countries with higher economic output tend to provide better healthcare,nutrition and living conditions,resulting in longer life expectancy.   

**insight**
GDP strongly influences life expectancy,especially at lower income llevels,but healthcare quality,education,lifestyle, and social factors also play a crucial role once basic economic needs are met.

***Average life expectancy by status***

Let's calculate average life expectancy by status

```python

# Calculate average life expectancy by status
status_avg = df.groupby('Status')['Life expectancy '].mean()

# Plot bar chart
plt.figure()
status_avg.plot(kind='bar')
plt.title('Average Life Expectancy by Country Status')
plt.xlabel('Country Status')
plt.ylabel('Average Life Expectancy')
plt.show()
```

The bar chart shows that developed countries have a higher average life expectancy compared to developing countries.

**Life expectancy by adult mortality level**

```python


# Create Adult Mortality levels
df['Adult_Mortality_Level'] = pd.cut(
    df['Adult Mortality'],
    bins=[0, 100, 200, 300, 400, 600],
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)

# Create boxplot
plt.figure()
sns.boxplot(
    x='Adult_Mortality_Level',
    y='Life expectancy ',
    data=df
)
plt.title('Life Expectancy by Adult Mortality Level')
plt.xlabel('Adult Mortality Level')
plt.ylabel('Life Expectancy')
plt.show()
```

the boxplot shows that life expectancy decreases as adult mortaliyty level increases.countries with very low adult mortality have higher median life expectancy and smaller variation,while very higher mortality levels show lower median life expectancy and greater variability.

**Correlation Heatmap**

   Correlation measures the strength and direction of the relationship between two numerical variables.

   The correlation coefficient ranges from -1 to +1:

   positive value - as years of experience increase,monthly salary tends to increase.

   Negative value - as one variable increases,the other tends to decrease.
   
   value near 0 - little or linear relationship.

Life expectancy is positively correlated with education and economic factors and negatively correlated with mortality and disease indicators.

```python
plt.figure(figsize=(15, 12))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

**Violinplot**

**Lineplot**

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df,
    x="Year",
    y="Life expectancy ",
    hue="Status"
)
plt.title("Life Expectancy Trend Over Years by Status")
plt.show()
```

**swarmplot**

```python
sns.swarmplot(
    x='Status',
    y='Life expectancy ',
    hue='Year',
    data=df,
    size=3
)

plt.title('Life Expectancy by Country Status Across Years')
plt.xlabel('Country Status')
plt.ylabel('Life Expectancy ')
plt.legend(title='Year')
plt.show()
```

The swarm plot shows that life expectancy has increased over the years for both country groups,but developed countries consistently maintain higher life expectancy than developed countries.

**Insight**

 although life expectancy has improved globally over time,developing countries continue tolag behaind developed countries, highlighting the need for stronger investments in healthcare,sanitation,nutrition and public health policies to close the gap.
